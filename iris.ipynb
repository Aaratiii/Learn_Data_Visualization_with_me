```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
iris_filepath = r"C:\Users\user\Desktop\cre\Data-Visualization\iris (1).csv" 
```

```python
iris = pd.read_csv(iris_filepath)
```

```python
iris.head()
```

```python
#Histogram

sns.histplot(iris['Petal Length (cm)'])
```

```python
#Density plots


# KDE plot 
sns.kdeplot(data=iris['Petal Length (cm)'], shade=True)
```

```python
sns.kdeplot(data=iris['Sepal Length (cm)'], shade=True)
```

```python
sns.kdeplot(data=iris['Petal Width (cm)'], shade=True)
```

```python
# 2D KDE plots

sns.jointplot(x=iris['Petal Length (cm)'], y=iris['Sepal Width (cm)'], kind="kde")

```

```python
#Color-coded plots


# Histograms for each species
sns.histplot(data=iris, x='Petal Length (cm)', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths, by Species")
```

```python
sns.kdeplot(data=iris, x = 'Petal Width (cm)', hue='Species', shade=True)
```
